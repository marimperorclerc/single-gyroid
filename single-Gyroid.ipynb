### intensities for $I4_132$ single gyroid shape

```python
import sgyroid as sg
```

volume fraction depending on t value
* t=0: full half gyroid: volume_fraction=0.5
* t= 1.5: upper limit: volume_fraction=0

```python
nval=8
for i in range(0,nval):
    tmin=0.
    tmax=1.4
    t=tmin+i*(tmax-tmin)/nval
    print("t=",t," volume fraction =",sg.volume_fraction(1,t,nmax=50))
```

```python
hkllist=[
    [[1,1,0],12],
    [[2,1,1],24],
    [[2,2,0],12],
    [[3,1,0],24],
    [[2,2,2],8],
    [[3,2,1],48],
    [[4,0,0],6]
]
```

```python
for hkl in hkllist:
    print("hkl=",hkl[0],"multiplicity=",hkl[1])
```

```python
t=0.6
nmax=50 #number of integration steps
print("powder Ihkl")
print("volume fraction = ",sg.volume_fraction(1,t,nmax))
norm=100000 #arbitrary coefficient to get large Ihkl values
for hkl in hkllist:
   h=hkl[0][0]
   k=hkl[0][1]
   l=hkl[0][2]      
   print("hkl=",hkl[0],"Ihkl",norm*hkl[1]*sg.Ihkl(1,t,h,k,l,nmax)/(h*h+k*k+l*l))
```

```python
nval=8
nmax=50 #number of integration steps
norm=100000 #arbitrary coefficient to get large Ihkl values

print("powder Ihkl")

print("nmax=",nmax)

for i in range(0,nval):
   tmin=0.
   tmax=1.4
   t=tmin+i*(tmax-tmin)/nval
   print("  ")
   print("volume fraction = ",sg.volume_fraction(1,t,nmax))
   for hkl in hkllist:
      h=hkl[0][0]
      k=hkl[0][1]
      l=hkl[0][2]      
      print("hkl=",hkl[0],"Ihkl",norm*hkl[1]*sg.Ihkl(1,t,h,k,l,nmax)/(h*h+k*k+l*l))
```
