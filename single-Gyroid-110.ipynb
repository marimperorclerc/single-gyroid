* 2D powder correction is a divison by $K_{perp}$, the perpendicular component of a Bragg reflection to the u direction (here along 011)
* it is not completely correct, because this perpendicular component may take different values for different hkl in the same family, like 321 and 312 ... but it gives some trend.

```python
t=0.6
nmax=50 #number of integration steps
print("powder Ihkl")
print("volume fraction = ",sg.volume_fraction(1,t,nmax))
norm=100000 #arbitrary coefficient to get large Ihkl values
for hkl in hkllist011:
   h=hkl[0][0]
   k=hkl[0][1]
   l=hkl[0][2]
   powder2D=h*h+(k-l)**2/2      
   print("hkl=",hkl[0],"Ihkl",norm*hkl[1]*sg.Ihkl(1,t,h,k,l,nmax)/powder2D)
```
